# step 4
### feature engineering

```python
def public_or_private(x):
    if x == 1:
        return 'Public'
    else:
        return 'Private'
```

```python
a = 1 
b = 2
```

```python
public_or_private(a)
```

```python
df['Public or Private'] = df['Public (1)/ Private (2)'].apply(public_or_private)
```

```python
df['total cost (in state tuition)'] = df['in-state tuition'] + df['room'] + df['board'] + df['add. fees'] + df['estim. book costs'] + df['estim. personal $']
    
```

```python
df['total cost (out of state tuition)'] = df['out-of-state tuition'] + df['room'] + df['board'] + df['add. fees'] + df['estim. book costs'] + df['estim. personal $']
    
```

```python
df['Mean Tuition'] = (df['in-state tuition'] + df['out-of-state tuition']) / 2 
```

```python
low_amount_x = df['Mean Tuition'].quantile(0.25)
up_amount_x = df['Mean Tuition'].quantile(0.75)
x_range_1 = up_amount_x - low_amount_x
lower_bound_x = low_amount_x - 1.5 * x_range_1
upper_bound_x = up_amount_x + 1.5 * x_range_1
```

```python
low_amount_x2 = df['stud./fac. ratio'].quantile(0.25)
up_amount_x2 = df['stud./fac. ratio'].quantile(0.75)
x_range_2 = up_amount_x2 - low_amount_x2
lower_bound_x2 = low_amount_x2 - 1.5 * x_range_2
upper_bound_x2 = up_amount_x2 + 1.5 * x_range_2
```

```python
low_amount_y = df['Graduation rate'].quantile(0.25)
up_amount_y = df['Graduation rate'].quantile(0.75)
y_range = up_amount_y - low_amount_y
lower_bound_y = low_amount_y - 1.5 * y_range
upper_bound_y = up_amount_y + 1.5 * y_range
```
